**Calculate profits and returns**

To understand profits and returns in the context of investment,I need to calculate

Absolute Return: The overall return on an investment without adjusting for risk or comparing to benchmarks.      

ROI (Return on Investment): A percentage that measures the return relative to the initial investment.

Annualized Return: The geometric average amount of money earned by an investment each year over a given time period.

```python
import yfinance as yf
import numpy as np
import datetime

# Function to fetch stock data
def get_stock_data(ticker, start_date, end_date):
    stock = yf.Ticker(ticker)
    data = stock.history(start=start_date, end=end_date)
    return data['Close']  # Use 'Close' for price data

# Function to calculate Absolute Return
def absolute_return(initial_price, final_price):
    return (final_price - initial_price) / initial_price * 100

# Function to calculate ROI
def roi(initial_investment, final_value):
    return (final_value - initial_investment) / initial_investment * 100

# Function to calculate Annualized Return
def annualized_return(initial_price, final_price, years):
    return (final_price / initial_price) ** (1 / years) - 1

# Define stock symbols
stocks = ["AAPL", "TSLA", "GOOGL"]  # Add more stocks as needed

# Define time period (1 year, 5 years, etc.)
start_date = "2018-02-01"
end_date = "2024-02-01"

# Analyze stocks
for stock in stocks:
    print(f"\n {stock} Performance Analysis:")

    # Fetch stock data
    data = get_stock_data(stock, start_date, end_date)

    # Get initial and final prices
    initial_price = data.iloc[0]
    final_price = data.iloc[-1]

    # Calculate Absolute Return
    abs_return = absolute_return(initial_price, final_price)
    print(f"  Absolute Return: {abs_return:.2f}%")

    # Calculate ROI
    initial_investment = 1000  # Example initial investment
    final_value = initial_investment * (final_price / initial_price)
    roi_percentage = roi(initial_investment, final_value)
    print(f"  ROI: {roi_percentage:.2f}%")

    # Calculate Annualized Return
    years = (data.index[-1] - data.index[0]).days / 365  # Convert days to years
    annualized_ret = annualized_return(initial_price, final_price, years)
    print(f"  Annualized Return: {annualized_ret*100:.2f}%")

```

```python
import yfinance as yf
import numpy as np

# Function to fetch historical stock data
def get_stock_data(ticker, start_date, end_date):
    stock = yf.Ticker(ticker)
    data = stock.history(start=start_date, end=end_date)
    return data['Close'].pct_change().dropna()  # Calculate daily returns

# Function to calculate Sharpe Ratio
def sharpe_ratio(returns, risk_free_rate=0.03):  # Risk-free rate ~3% (adjustable)
    mean_return = np.mean(returns) * 252  # Annualized return
    std_dev = np.std(returns) * np.sqrt(252)  # Annualized volatility
    return (mean_return - risk_free_rate) / std_dev if std_dev != 0 else None

# Define stock symbol and benchmarks
stock_ticker = "AAPL"  # Example: Apple stock
benchmarks = {
    "S&P 500": "^GSPC",
    "NASDAQ": "^IXIC",
    "Dow Jones": "^DJI"
}

# Define timeframes
timeframes = {
    "1-Year": ("2023-02-01", "2024-02-01"),
    "5-Year": ("2019-02-01", "2024-02-01"),
    "10-Year": ("2014-02-01", "2024-02-01"),
}

# Get Sharpe Ratios for stock
stock_sharpe_ratios = {period: sharpe_ratio(get_stock_data(stock_ticker, start, end))
                        for period, (start, end) in timeframes.items()}

# Get Sharpe Ratios for benchmarks
benchmark_sharpe_ratios = {name: {period: sharpe_ratio(get_stock_data(ticker, start, end))
                                   for period, (start, end) in timeframes.items()}
                           for name, ticker in benchmarks.items()}

# Display results
print(f"Stock: {stock_ticker}\n")

# Print Stock Sharpe Ratios
for period, ratio in stock_sharpe_ratios.items():
    print(f"{period} Sharpe Ratio: {ratio:.2f}")

print("\nMarket Benchmarks:")
# Print Benchmark Sharpe Ratios
for name, ratios in benchmark_sharpe_ratios.items():
    print(f"\n{name}:")
    for period, ratio in ratios.items():
        print(f"{period} Sharpe Ratio: {ratio:.2f}")

# Compare Risk-Adjusted Performance
print("\nRisk-Adjusted Performance Analysis:")
for period in timeframes.keys():
    stock_ratio = stock_sharpe_ratios.get(period, None)
    if stock_ratio is None:
        continue

    print(f"\n {period} Performance:")
    for name, ratios in benchmark_sharpe_ratios.items():
        benchmark_ratio = ratios.get(period, None)
        if benchmark_ratio is None:
            continue

        if stock_ratio > benchmark_ratio:
            print(f" {stock_ticker} had a **better** risk-adjusted return than {name} ({stock_ratio:.2f} vs. {benchmark_ratio:.2f})")
        else:
            print(f" {stock_ticker} had a **worse** risk-adjusted return than {name} ({stock_ratio:.2f} vs. {benchmark_ratio:.2f})")

# Interpretation
print("\n Interpretation Guide:")
print(" Sharpe Ratio > 1 → Good risk-adjusted return")
print(" Sharpe Ratio < 1 → Too much risk for the return")
```

```python
import yfinance as yf
import numpy as np
import datetime

# Function to fetch historical data for a stock or benchmark
def get_data(ticker, start_date, end_date):
    data = yf.Ticker(ticker)
    return data.history(start=start_date, end=end_date)['Close']

# Function to calculate Absolute Return
def absolute_return(initial_price, final_price):
    return (final_price - initial_price) / initial_price * 100

# Function to calculate ROI
def roi(initial_investment, initial_price, final_price):
    final_value = initial_investment * (final_price / initial_price)
    return (final_value - initial_investment) / initial_investment * 100

# Function to calculate Sharpe Ratio (Risk-adjusted Return)
def sharpe_ratio(returns, risk_free_rate=0.02):
    excess_returns = returns - risk_free_rate / 252  # Daily risk-free rate (assuming 252 trading days)
    return np.mean(excess_returns) / np.std(excess_returns)  # Mean excess return over standard deviation

# Define stock and benchmarks
stock_ticker = "AAPL"
benchmarks = {
    "S&P 500": "^GSPC",
    "NASDAQ": "^IXIC",
    "Dow Jones": "^DJI"
}

# Define time period (1 year, 5 years, etc.)
start_date = "2018-02-01"
end_date = "2024-02-01"

# Fetch stock data
stock_data = get_data(stock_ticker, start_date, end_date)
stock_initial_price = stock_data.iloc[0]
stock_final_price = stock_data.iloc[-1]

# Compare for each benchmark
for benchmark_name, benchmark_ticker in benchmarks.items():
    print(f"\n {benchmark_name} Performance Comparison with {stock_ticker}:")

    # Fetch benchmark data
    benchmark_data = get_data(benchmark_ticker, start_date, end_date)
    benchmark_initial_price = benchmark_data.iloc[0]
    benchmark_final_price = benchmark_data.iloc[-1]

    # Calculate ROI for stock and benchmark
    stock_roi = roi(1000, stock_initial_price, stock_final_price)
    benchmark_roi = roi(1000, benchmark_initial_price, benchmark_final_price)

    # Calculate Absolute Return for stock and benchmark
    stock_abs_return = absolute_return(stock_initial_price, stock_final_price)
    benchmark_abs_return = absolute_return(benchmark_initial_price, benchmark_final_price)

    # Calculate Daily Returns for Sharpe Ratio
    stock_returns = stock_data.pct_change().dropna()
    benchmark_returns = benchmark_data.pct_change().dropna()

    # Calculate Sharpe Ratio for stock and benchmark
    stock_sharpe = sharpe_ratio(stock_returns)
    benchmark_sharpe = sharpe_ratio(benchmark_returns)

    # Print comparisons
    print(f"  ROI: Stock = {stock_roi:.2f}%, Benchmark = {benchmark_roi:.2f}%")
    if stock_roi > benchmark_roi:
        print("   Stock outperformed the Benchmark in ROI.")
    else:
        print("   Stock underperformed the Benchmark in ROI.")

    print(f"  Absolute Return: Stock = {stock_abs_return:.2f}%, Benchmark = {benchmark_abs_return:.2f}%")
    if stock_abs_return > benchmark_abs_return:
        print("   Stock outperformed the Benchmark in Absolute Return.")
    else:
        print("   Stock underperformed the Benchmark in Absolute Return.")

    print(f"  Sharpe Ratio: Stock = {stock_sharpe:.2f}, Benchmark = {benchmark_sharpe:.2f}")
    if stock_sharpe > benchmark_sharpe:
        print("   Stock has a better risk-adjusted return (Sharpe Ratio).")
    else:
        print("   Stock has a lower risk-adjusted return (Sharpe Ratio).")

```

**Risk Identification & Financial Metrics**

```python
# Function to calculate Absolute Return
def absolute_return(initial_price, final_price):
    return (final_price - initial_price) / initial_price * 100

# Function to calculate ROI
def roi(initial_investment, initial_price, final_price):
    final_value = initial_investment * (final_price / initial_price)
    return (final_value - initial_investment) / initial_investment * 100

# Function to calculate Sharpe Ratio (Risk-adjusted Return)
def sharpe_ratio(returns, risk_free_rate=0.05):
    excess_returns = returns - risk_free_rate / 500  # Daily risk-free rate (assuming 500 trading days)
    return np.mean(excess_returns) / np.std(excess_returns)  # Mean excess return over standard deviation

# Function to calculate Value at Risk (VaR) at a given confidence level (e.g., 95%)
def value_at_risk(returns, confidence_level=0.95):
    return np.percentile(returns, 100 * (1 - confidence_level))

# Function to calculate Conditional Value at Risk (CVaR) (expected shortfall)
def conditional_value_at_risk(returns, confidence_level=0.95):
    var_threshold = value_at_risk(returns, confidence_level)
    return np.mean(returns[returns < var_threshold])

# Define stock and benchmarks
stock_ticker = "AAPL"
benchmarks = {
    "S&P 500": "^GSPC",
    "NASDAQ": "^IXIC",
    "Dow Jones": "^DJI"
}

# Define time period (1 year, 5 years, etc.)
start_date = "2018-02-01"
end_date = "2024-02-01"

# Fetch stock data
stock_data = get_data(stock_ticker, start_date, end_date)
stock_initial_price = stock_data.iloc[0]
stock_final_price = stock_data.iloc[-1]

# Compare for each benchmark
for benchmark_name, benchmark_ticker in benchmarks.items():
    print(f"\n {benchmark_name} Performance & Risk Analysis for {stock_ticker}:")

    # Fetch benchmark data
    benchmark_data = get_data(benchmark_ticker, start_date, end_date)
    benchmark_initial_price = benchmark_data.iloc[0]
    benchmark_final_price = benchmark_data.iloc[-1]

    # Calculate ROI for stock and benchmark
    stock_roi = roi(1000, stock_initial_price, stock_final_price)
    benchmark_roi = roi(1000, benchmark_initial_price, benchmark_final_price)

    # Calculate Absolute Return for stock and benchmark
    stock_abs_return = absolute_return(stock_initial_price, stock_final_price)
    benchmark_abs_return = absolute_return(benchmark_initial_price, benchmark_final_price)

    # Calculate Daily Returns for Risk Analysis
    stock_returns = stock_data.pct_change().dropna()
    benchmark_returns = benchmark_data.pct_change().dropna()

    # Calculate Sharpe Ratio for stock and benchmark
    stock_sharpe = sharpe_ratio(stock_returns)
    benchmark_sharpe = sharpe_ratio(benchmark_returns)

    # Calculate VaR and CVaR for stock and benchmark
    stock_var_95 = value_at_risk(stock_returns)
    stock_cvar_95 = conditional_value_at_risk(stock_returns)
    benchmark_var_95 = value_at_risk(benchmark_returns)
    benchmark_cvar_95 = conditional_value_at_risk(benchmark_returns)

    # Print comparisons
    print(f"  ROI: Stock = {stock_roi:.2f}%, Benchmark = {benchmark_roi:.2f}%")
    print(f"  Absolute Return: Stock = {stock_abs_return:.2f}%, Benchmark = {benchmark_abs_return:.2f}%")
    print(f"  Sharpe Ratio: Stock = {stock_sharpe:.2f}, Benchmark = {benchmark_sharpe:.2f}")

    print("\n Risk Metrics:")
    print(f"  Value at Risk (VaR 95%): Stock = {stock_var_95:.2%}, Benchmark = {benchmark_var_95:.2%}")
    print(f"  Conditional Value at Risk (CVaR 95%): Stock = {stock_cvar_95:.2%}, Benchmark = {benchmark_cvar_95:.2%}")

    if stock_var_95 > benchmark_var_95:
        print("  Stock is more volatile than the benchmark.")
    else:
        print(" Stock has lower downside risk compared to the benchmark.")

    if stock_cvar_95 > benchmark_cvar_95:
        print("  Stock has a higher expected loss beyond VaR compared to the benchmark.")
    else:
        print("  Stock has a lower tail risk compared to the benchmark.")

# Risk Identification Summary
print("\n **Risk Identification in Stock Trading:**")
print("- **Market Crashes:** 2008 Financial Crisis, COVID-19 Market Crash (March 2020).")
print("- **Liquidity Risks:** Stocks with high bid-ask spreads and low trading volume can be harder to buy/sell.")
print("- **Technical Risks:** Flash crashes, algorithmic trading failures, and system errors can cause unexpected price drops.")
print("- **Quantitative Risk Metrics:** Value at Risk (VaR) and Conditional Value at Risk (CVaR) measure potential loss in worst-case scenarios.")
```

**Risk Management & Basel III Compliance**

```python
import numpy as np

# Basel III Risk Management Metrics

# Function to calculate Capital Adequacy Ratio (CAR)
def capital_adequacy_ratio(tier_1_capital, tier_2_capital, risk_weighted_assets):
    """
    CAR = (Tier 1 Capital + Tier 2 Capital) / Risk-Weighted Assets
    Basel III minimum requirement: 10.5% (including capital conservation buffer)
    """
    return ((tier_1_capital + tier_2_capital) / risk_weighted_assets) * 100

# Function to calculate Leverage Ratio
def leverage_ratio(tier_1_capital, total_exposure):
    """
    Leverage Ratio = Tier 1 Capital / Total Exposure
    Basel III minimum requirement: 3%
    """
    return (tier_1_capital / total_exposure) * 100

# Function to calculate Liquidity Coverage Ratio (LCR)
def liquidity_coverage_ratio(high_quality_liquid_assets, net_cash_outflows_30_days):
    """
    LCR = High-Quality Liquid Assets / Net Cash Outflows over 30 days
    Basel III minimum requirement: 100%
    """
    return (high_quality_liquid_assets / net_cash_outflows_30_days) * 100

# Sample Data (Replace with real banking data)
tier_1_capital = 500  # Core equity capital (in million USD)
tier_2_capital = 200  # Supplementary capital (in million USD)
risk_weighted_assets = 4000  # Total risk-weighted assets (in million USD)
total_exposure = 12000  # Total assets and off-balance exposures (in million USD)
high_quality_liquid_assets = 800  # Liquid assets available (in million USD)
net_cash_outflows_30_days = 600  # Expected cash outflows over 30 days (in million USD)

# Calculate Basel III Metrics
car = capital_adequacy_ratio(tier_1_capital, tier_2_capital, risk_weighted_assets)
leverage = leverage_ratio(tier_1_capital, total_exposure)
lcr = liquidity_coverage_ratio(high_quality_liquid_assets, net_cash_outflows_30_days)

# Print Basel III Compliance Metrics
print("\n **Basel III Compliance Metrics:**")
print(f"  Capital Adequacy Ratio (CAR) = {car:.2f}%")
print(f"  Leverage Ratio = {leverage:.2f}%")
print(f"  Liquidity Coverage Ratio (LCR) = {lcr:.2f}%")

# Basel III Compliance Check
print("\n **Compliance Check:**")
if car >= 10.5:
    print("  CAR meets Basel III minimum requirement (10.5%)")
else:
    print("   CAR is below Basel III requirement!")

if leverage >= 3:
    print("   Leverage Ratio meets Basel III minimum requirement (3%)")
else:
    print("   Leverage Ratio is below Basel III requirement!")

if lcr >= 100:
    print("   Liquidity Coverage Ratio meets Basel III minimum requirement (100%)")
else:
    print("   Liquidity Coverage Ratio is below Basel III requirement!")
```

```python
import yfinance as yf
import numpy as np

# Function to fetch historical stock data
def get_stock_data(ticker, start_date, end_date):
    stock = yf.Ticker(ticker)
    data = stock.history(start=start_date, end=end_date)['Close']
    return data

# Function to calculate Value at Risk (VaR)
def value_at_risk(returns, confidence_level=0.95):
    """
    VaR estimates the maximum expected loss at a given confidence level.
    Example: If VaR = -3%, there's a 95% chance the stock won't lose more than 3% in a day.
    """
    return np.percentile(returns, (1 - confidence_level) * 100)

# Function to calculate Conditional VaR (CVaR)
def conditional_var(returns, confidence_level=0.95):
    """
    CVaR estimates the expected loss beyond the VaR threshold.
    Example: If VaR = -3%, CVaR calculates the average loss beyond -3%.
    """
    var_threshold = value_at_risk(returns, confidence_level)
    return returns[returns <= var_threshold].mean()

# Define stock and time period
stock_ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2024-02-01"

# Fetch stock data
stock_data = get_stock_data(stock_ticker, start_date, end_date)
stock_returns = stock_data.pct_change().dropna()

# Calculate risk metrics
var_95 = value_at_risk(stock_returns, confidence_level=0.95)
cvar_95 = conditional_var(stock_returns, confidence_level=0.95)

# Print Financial Risk Report
print("\n **Financial Risk Exposure Report:**")
print(f"  Stock: {stock_ticker}")
print(f"  Value at Risk (VaR 95%): {var_95:.2%}")
print(f"  Conditional VaR (CVaR 95%): {cvar_95:.2%}")

# Risk Analysis
print("\n **Risk Analysis:**")
if var_95 > -0.02:
    print("   Stock has low risk (VaR < -2%).")
else:
    print("   Stock has high risk (VaR > -2%).")

if cvar_95 > -0.03:
    print("   Stock is relatively stable (CVaR < -3%).")
else:
    print("   Stock has extreme downside risk (CVaR > -3%).")
```

**Risk Metric Comparison**

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Function to fetch historical data
def get_stock_data(ticker, start_date, end_date):
    data = yf.Ticker(ticker).history(start=start_date, end=end_date)
    return data['Close'].pct_change().dropna()  # Daily returns

# Function to calculate Value at Risk (VaR) at a given confidence level
def calculate_var(returns, confidence_level=0.95):
    return np.percentile(returns, (1 - confidence_level) * 100)

# Function to calculate Conditional Value at Risk (CVaR)
def calculate_cvar(returns, confidence_level=0.95):
    var_threshold = calculate_var(returns, confidence_level)
    return returns[returns <= var_threshold].mean()

# Function to calculate Sharpe Ratio
def calculate_sharpe_ratio(returns, risk_free_rate=0.02):
    excess_returns = returns - (risk_free_rate / 252)  # Daily risk-free rate
    return np.mean(excess_returns) / np.std(returns)

# Function to calculate Sortino Ratio
def calculate_sortino_ratio(returns, risk_free_rate=0.02):
    downside_returns = returns[returns < 0]  # Focus only on negative returns
    excess_returns = returns - (risk_free_rate / 252)
    return np.mean(excess_returns) / np.std(downside_returns)

# Define stocks and benchmarks
stocks = {"AAPL": "AAPL", "S&P 500": "^GSPC", "NASDAQ": "^IXIC", "Dow Jones": "^DJI"}

# Define time period
start_date = "2019-01-01"
end_date = "2024-02-01"

# Store risk metrics results
risk_metrics = {}

# Calculate risk metrics for each asset
for name, ticker in stocks.items():
    returns = get_stock_data(ticker, start_date, end_date)

    var_95 = calculate_var(returns, 0.95)
    cvar_95 = calculate_cvar(returns, 0.95)
    sharpe = calculate_sharpe_ratio(returns)
    sortino = calculate_sortino_ratio(returns)

    risk_metrics[name] = {
        "VaR 95%": var_95,
        "CVaR 95%": cvar_95,
        "Sharpe Ratio": sharpe,
        "Sortino Ratio": sortino
    }

# Convert results to a DataFrame for better visualization
risk_df = pd.DataFrame(risk_metrics).T
print("\nRisk Metric Comparison:")
print(risk_df)

# Plot risk-adjusted performance
risk_df[["Sharpe Ratio", "Sortino Ratio"]].plot(kind="bar", figsize=(10, 5), title="Risk-Adjusted Performance Comparison")
plt.ylabel("Ratio")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
