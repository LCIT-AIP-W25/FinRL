```python
import os
import pandas as pd

def list_csv_files(directory):
    """
    List all CSV files in the given directory and display their names.
    
    Args:
        directory (str): The path to the directory containing the CSV files.
        
    Returns:
        list: A list of CSV file names.
    """
    csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]
    return csv_files

def load_and_combine_data(directory):
    """
    Load all CSV files from the given directory and combine them into a single DataFrame.
    
    Args:
        directory (str): The path to the directory containing the CSV files.
        
    Returns:
        pd.DataFrame: A combined DataFrame of all CSV files.
    """
    combined_df = pd.DataFrame()
    csv_files = list_csv_files(directory)
    for file in csv_files:
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path)
        #df['FileName'] = file  # Add a column for the CSV file name
        df['Stock'] = file.replace('.csv', '')  # Add a column for the stock symbol
        combined_df = pd.concat([combined_df, df], ignore_index=True)
    return combined_df

# Example usage
directory_path = "C:/Users/jini/Desktop/AIP_documents/archive (9)/stock_market_data/sp500/csv"
csv_files = list_csv_files(directory_path)

print("CSV Files:")
for file in csv_files:
    print(file)

combined_data = load_and_combine_data(directory_path)

# Display the first few rows of the combined data
print("\nCombined Data:")
print(combined_data.head())
print(combined_data.tail())
```

```python
# Check the structure of the combined data
print("Data Info:")
print(combined_data.info())

# Check for missing values
print("\nMissing Values:")
print(combined_data.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(combined_data.describe())

# Check unique stocks in the dataset
print("\nUnique Stocks:")
print(combined_data['Stock'].unique())

# Check date range
print("\nDate Range:")
print("Start Date:", combined_data['Date'].min())
print("End Date:", combined_data['Date'].max())
```

```python
# Convert 'Date' column to datetime format
combined_data['Date'] = pd.to_datetime(combined_data['Date'], format='%d-%m-%Y')

# Sort data by 'Stock' and 'Date'
combined_data = combined_data.sort_values(by=['Stock', 'Date'])

# Calculate daily returns for each stock
combined_data['Daily_Return'] = combined_data.groupby('Stock')['Close'].pct_change()

# Drop rows with missing values (e.g., first row for each stock will have NaN for Daily_Return)
combined_data = combined_data.dropna()

# Display the first few rows after preprocessing
print("\nPreprocessed Data:")
print(combined_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot daily returns for a specific stock
stock_A_data = combined_data[combined_data['Stock'] == 'A']
plt.figure(figsize=(10, 6))
plt.plot(stock_A_data['Date'], stock_A_data['Daily_Return'], label='Daily Return')
plt.title('Daily Returns for Stock A')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()

# Distribution of daily returns
plt.figure(figsize=(10, 6))
sns.histplot(stock_A_data['Daily_Return'], kde=True, bins=50)
plt.title('Distribution of Daily Returns for Stock A')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()

# Correlation matrix for numerical features
correlation_matrix = stock_A_data[['Low', 'Open', 'High', 'Close', 'Volume', 'Daily_Return']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Stock A')
plt.show()
```

```python
# Calculate Moving Averages
combined_data['MA_10'] = combined_data.groupby('Stock')['Close'].transform(lambda x: x.rolling(window=10).mean())
combined_data['MA_50'] = combined_data.groupby('Stock')['Close'].transform(lambda x: x.rolling(window=50).mean())

# Calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

combined_data['RSI'] = combined_data.groupby('Stock').apply(lambda x: calculate_rsi(x)).reset_index(level=0, drop=True)

# Calculate Volatility (Standard Deviation of Daily Returns)
combined_data['Volatility'] = combined_data.groupby('Stock')['Daily_Return'].transform(lambda x: x.rolling(window=10).std())

# Drop rows with missing values (due to rolling calculations)
combined_data = combined_data.dropna()

# Display the first few rows after feature engineering
print("\nData After Feature Engineering:")
print(combined_data.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize features
scaler = MinMaxScaler()
features_to_scale = ['Close', 'Volume', 'MA_10', 'MA_50', 'RSI', 'Volatility']
combined_data[features_to_scale] = scaler.fit_transform(combined_data[features_to_scale])

# Display the first few rows after normalization
print("\nData After Normalization:")
print(combined_data.head())
```

```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed

# Define the calculate_var_cvar function in the global scope
def calculate_var_cvar(returns, confidence_level=0.95):
    """
    Calculate Value at Risk (VaR) and Conditional Value at Risk (CVaR).
    
    Args:
        returns (pd.Series): Historical returns of the portfolio.
        confidence_level (float): Confidence level for VaR and CVaR (e.g., 0.95 for 95%).
        
    Returns:
        float: VaR and CVaR values.
    """
    # Sort returns in ascending order
    returns_sorted = returns.sort_values(ascending=True)
    
    # Calculate VaR as the (1 - confidence_level) quantile
    var = returns_sorted.quantile(1 - confidence_level)
    
    # Calculate CVaR as the mean of returns worse than VaR
    cvar = returns_sorted[returns_sorted <= var].mean()
    
    return var, cvar

# Function to calculate metrics for a single stock
def calculate_metrics_for_stock(stock_data):
    stock = stock_data['Stock'].iloc[0]
    returns = stock_data['Daily_Return']
    var, cvar = calculate_var_cvar(returns, confidence_level=0.95)
    return stock, var, cvar
```

```python
from joblib import Parallel, delayed

# Parallelize the calculation for all stocks
results = Parallel(n_jobs=-1)(delayed(calculate_metrics_for_stock)(group) for name, group in grouped_data)

# Convert results to a DataFrame
risk_metrics_df = pd.DataFrame(results, columns=['Stock', 'VaR', 'CVaR'])
print(risk_metrics_df)
```
