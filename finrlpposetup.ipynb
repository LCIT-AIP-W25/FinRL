```python
import gym
import numpy as np
import pandas as pd
from gym import spaces
from gym.utils import seeding

class StockTradingEnv(gym.Env):
    def __init__(self, data, window_size=50):
        super(StockTradingEnv, self).__init__()
        
        # Initialize the stock data and window size
        self.data = data
        self.window_size = window_size
        
        # Define action space: 0 = hold, 1 = buy, 2 = sell
        self.action_space = spaces.Discrete(3)  # [hold, buy, sell]
        
        # Define observation space: a window of 'window_size' previous data points
        # The observation will contain (price, technical indicators)
        self.observation_space = spaces.Box(low=0, high=1, shape=(window_size, len(data.columns)), dtype=np.float32)
        
        self.current_step = 0
        self.done = False
        self.balance = 10000  # Starting with $10,000 balance
        self.stock_held = 0
        self.initial_balance = 10000
        self.max_balance = 0
        
    def reset(self):
        # Reset the state of the environment
        self.current_step = self.window_size
        self.balance = self.initial_balance
        self.stock_held = 0
        self.done = False
        self.max_balance = self.initial_balance
        return self._next_observation()
    
    def _next_observation(self):
        # Return the current window of data
        window_data = self.data.iloc[self.current_step - self.window_size:self.current_step]
        return np.array(window_data)
    
    def step(self, action):
        # Take an action (buy, sell, hold) and update the environment
        current_price = self.data.iloc[self.current_step]['Close/Last']
        previous_balance = self.balance + self.stock_held * current_price
        
        if action == 1:  # Buy
            if self.balance >= current_price:  # If we have enough balance
                self.stock_held += 1
                self.balance -= current_price
                
        elif action == 2:  # Sell
            if self.stock_held > 0:  # If we hold stocks
                self.stock_held -= 1
                self.balance += current_price
        
        # Update max balance
        total_balance = self.balance + self.stock_held * current_price
        self.max_balance = max(self.max_balance, total_balance)
        
        # Move to the next step
        self.current_step += 1
        
        # Check if we've reached the end of the data
        if self.current_step >= len(self.data) - 1:
            self.done = True
        
        # Get the next observation
        next_observation = self._next_observation()
        
        # Calculate the reward (profit or loss)
        reward = total_balance - previous_balance
        
        return next_observation, reward, self.done, {}
    
    def render(self):
        # Render the current state of the environment
        print(f"Step: {self.current_step}")
        print(f"Balance: {self.balance}")
        print(f"Stock Held: {self.stock_held}")
        print(f"Total Balance: {self.balance + self.stock_held * self.data.iloc[self.current_step]['Close/Last']}")
```

```python
import gym
from gym import spaces
import numpy as np
import pandas as pd
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv

# Define the StockTradingEnv class
class StockTradingEnv(gym.Env):
    def __init__(self, data):
        super(StockTradingEnv, self).__init__()

        self.data = data
        self.n_steps = len(data)
        self.current_step = 0

        # Define action space: 3 discrete actions (buy, hold, sell)
        self.action_space = spaces.Discrete(3)

        # Define observation space: 4 features (Open, High, Low, Close)
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(4,), dtype=np.float32)

    def reset(self):
        self.current_step = 0
        self.done = False
        return self.data.iloc[self.current_step][['Open', 'High', 'Low', 'Close/Last']].values

    def step(self, action):
        self.current_step += 1

        if self.current_step >= self.n_steps - 1:
            self.done = True
        else:
            self.done = False

        current_data = self.data.iloc[self.current_step]
        obs = current_data[['Open', 'High', 'Low', 'Close/Last']].values

        # Reward logic: Buy = 1, Sell = -1, Hold = 0
        reward = 0
        if action == 0:  # Buy
            reward = 1
        elif action == 1:  # Sell
            reward = -1
        # Hold action is implicitly rewarded by zero.

        return obs, reward, self.done, {}

# Load and preprocess your data
data = pd.read_csv("C:/Users/jasha/Downloads/HistoricalData_1738851979485.csv")
data = data[['Date', 'Close/Last', 'Open', 'High', 'Low']]  # Simplify the data for now

# Wrap the environment in DummyVecEnv
env = DummyVecEnv([lambda: StockTradingEnv(data)])

# Create the PPO model (with a simple MLP policy)
model = PPO("MlpPolicy", env, verbose=1)

# Train the model (during training, you will see logs)
model.learn(total_timesteps=10000)
```
