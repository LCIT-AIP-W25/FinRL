**Calculation of Absolute Return and ROI**

Absolute return: Absolute return in stock trading is the total percentage gain or loss on an investment over a specific period

ROI (Return on Investment) measures the profitability of an investment relative to its cost.










```python
# Function to calculate Absolute Return and ROI
def calculate_absolute_return_roi(df):
    """
    Calculate Absolute Return and ROI for a given stock.

    :param df: DataFrame with 'Close' price column.
    :return: Dictionary containing Absolute Return and ROI.
    """
    if 'Close' not in df.columns:
        raise ValueError("Ensure the dataset has a 'Close' column.")

    # Get the initial and final closing prices
    initial_price = df['Close'].iloc[0]
    final_price = df['Close'].iloc[-1]

    # Calculate Absolute Return
    absolute_return = final_price - initial_price

    # Calculate ROI
    roi = ((final_price - initial_price) / initial_price) * 100

    return {
        'Absolute Return': absolute_return,
        'ROI (%)': roi
    }

# Analyze each stock and store results
results = []
for stock_file in stocks_to_analyze:
    # Load the stock data from the CSV file
    stock_path = os.path.join(dir_path, stock_file) # Corrected path using os.path.join

    # Check if file exists
    if not os.path.isfile(stock_path):
        print(f"Warning: File {stock_path} does not exist or is not a file. Skipping...")
        continue

    stock_data = pd.read_csv(stock_path) # Read the CSV file using the corrected path

    # Ensure the DataFrame has the required columns
    required_columns = ['Date', 'Close']
    if not all(column in stock_data.columns for column in required_columns):
        raise ValueError(f"CSV file {stock_file} must contain 'Date' and 'Close' columns.")

    # Convert 'Date' column to datetime format
    stock_data['Date'] = pd.to_datetime(stock_data['Date'], format='%d-%m-%Y', errors='coerce')

    # Sort by date (if not already sorted)
    stock_data = stock_data.sort_values(by='Date')

    # Extract the stock ticker from the file name
    ticker = stock_file.split('.')[0]

    # Calculate Absolute Return and ROI
    metrics = calculate_absolute_return_roi(stock_data)
    metrics['Ticker'] = ticker

    # Append results
    results.append(metrics)

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Display the results
print("Absolute Return and ROI for Each Stock:")
print(results_df)
```

```python
def calculate_metrics(df, benchmark_df):
    # Ensure price columns exist
    assert 'Close' in df.columns, "Ensure the dataset has a 'Close' column"
    assert 'Close' in benchmark_df.columns, "Ensure the benchmark dataset has a 'Close' column"

    # Calculate daily returns
    df['Daily Return'] = df['Close'].pct_change()
    benchmark_df['Benchmark Return'] = benchmark_df['Close'].pct_change()

    # Cumulative returns
    df['Cumulative Return'] = (1 + df['Daily Return']).cumprod()
    benchmark_df['Cumulative Benchmark Return'] = (1 + benchmark_df['Benchmark Return']).cumprod()

    # ROI Calculation
    initial_value = df['Close'].iloc[0]
    final_value = df['Close'].iloc[-1]
    ROI = (final_value - initial_value) / initial_value

    # Absolute Return Calculation
    absolute_return = final_value - initial_value

    # Sharpe Ratio (Assuming risk-free rate is 0)
    sharpe_ratio = df['Daily Return'].mean() / df['Daily Return'].std()

    # Maximum Drawdown
    rolling_max = df['Close'].cummax()
    drawdown = (df['Close'] - rolling_max) / rolling_max
    max_drawdown = drawdown.min()

    # Compare against benchmark
    benchmark_ROI = (benchmark_df['Close'].iloc[-1] - benchmark_df['Close'].iloc[0]) / benchmark_df['Close'].iloc[0]

    return {
        'ROI': ROI,
        'Absolute Return': absolute_return,
        'Benchmark ROI': benchmark_ROI,
        'Sharpe Ratio': sharpe_ratio,
        'Max Drawdown': max_drawdown,
        'Above Market': ROI > benchmark_ROI
    }

# Compute metrics
metrics = calculate_metrics(df, benchmark_df)
print(metrics)

# Extract values correctly for printing
roi = metrics['ROI']
absolute_return = metrics['Absolute Return']
sharpe_ratio = metrics['Sharpe Ratio']
benchmark_comparison = metrics['Above Market']

# Print formatted results
print(f"ROI: {roi:.2%}")
print(f"Absolute Return: ${absolute_return:.2f}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print("Benchmark Comparison:", benchmark_comparison)

# Compute risks
risks = identify_risks(df)
print(risks)

# Map out potential risks
potential_risks = map_out_risks()
print(potential_risks)
```

```python
import pandas as pd
import numpy as np
# Define the stocks to analyze (assuming each CSV file represents a stock)
stocks_to_analyze = ["AAPL.csv", "MSFT.csv", "GOOGL.csv"]  # Example: Apple, Microsoft, Google

# Function to calculate risk metrics for a single stock
def calculate_risk_metrics(stock_data):
    # Daily Returns
    stock_data['Daily_Return'] = stock_data['Close'].pct_change()

    # Volatility (Annualized)
    volatility = stock_data['Daily_Return'].std() * np.sqrt(252)

    # Maximum Drawdown
    stock_data['Cumulative_Return'] = (1 + stock_data['Daily_Return']).cumprod()
    peak = stock_data['Cumulative_Return'].cummax()
    drawdown = (stock_data['Cumulative_Return'] - peak) / peak
    max_drawdown = drawdown.min()

    # Liquidity (Average Daily Trading Volume)
    average_volume = stock_data['Volume'].mean()

    # Sharpe Ratio (Risk-Adjusted Return)
    risk_free_rate = 0.02  # Assume 2% risk-free rate
    excess_returns = stock_data['Daily_Return'] - (risk_free_rate / 252)
    sharpe_ratio = (excess_returns.mean() / excess_returns.std()) * np.sqrt(252)

    return {
        'Volatility': volatility,
        'Max_Drawdown': max_drawdown,
        'Average_Volume': average_volume,
        'Sharpe_Ratio': sharpe_ratio
    }

# Analyze each stock and store results
risk_metrics_comparison = []
for stock_file in stocks_to_analyze:
    # Load the stock data from the CSV file
    stock_path = os.path.join(dir_path, stock_file)
      # Check if the file exists before trying to read it
    if not os.path.exists(stock_path):
        print(f"Warning: File not found: {stock_path}. Skipping...")
        continue  # Skip to the next file

    stock_data = pd.read_csv(stock_path)

    # Ensure the DataFrame has the required columns
    required_columns = ['Date', 'Open', 'Close', 'Volume']
    if not all(column in stock_data.columns for column in required_columns):
        raise ValueError(f"CSV file {stock_file} must contain 'Date', 'Open', 'Close', and 'Volume' columns.")

    # Convert 'Date' column to datetime format
    stock_data['Date'] = pd.to_datetime(stock_data['Date'], format='%d-%m-%Y', errors='coerce')

    # Sort by date (if not already sorted)
    stock_data = stock_data.sort_values(by='Date')

    # Extract the stock ticker from the file name
    ticker = stock_file.split('.')[0]

    # Calculate risk metrics
    metrics = calculate_risk_metrics(stock_data)
    metrics['Ticker'] = ticker

    # Append results
    risk_metrics_comparison.append(metrics)

# Convert results to a DataFrame
risk_metrics_df = pd.DataFrame(risk_metrics_comparison)

# Display the results
print("Risk Metric Comparison:")
print(risk_metrics_df)

# Save the results to a CSV file
risk_metrics_df.to_csv('/content/drive/MyDrive/sp500_csv/risk_metrics_comparison.csv', index=False)
```

Calculation of Risk (VaR), Conditional Value at Risk
(CVaR)

```python
def calculate_var_cvar(df, confidence_level=0.95):

    if 'Close' not in df.columns:
        raise ValueError("Ensure the dataset has a 'Close' column.")

    # Calculate daily returns
    df['Daily Return'] = df['Close'].pct_change().dropna()

    # Compute Value at Risk (VaR)
    var_threshold = np.percentile(df['Daily Return'].dropna(), (1 - confidence_level) * 100)

    # Compute Conditional Value at Risk (CVaR)
    cvar = df['Daily Return'][df['Daily Return'] <= var_threshold].mean()

    # Additional Risk Metrics
    volatility = df['Daily Return'].std()  # Standard deviation as a measure of risk
    max_drawdown = (df['Close'] / df['Close'].cummax() - 1).min()  # Worst peak-to-trough drop

    return {
        'Value at Risk (VaR)': var_threshold,
        'Conditional Value at Risk (CVaR)': cvar,
        'Volatility': volatility,
        'Max Drawdown': max_drawdown
    }

# Compute risk metrics
risk_metrics = calculate_var_cvar(df)
print(risk_metrics)
```
