```python
import pandas as pd
import numpy as np

# Assuming `df` is your DataFrame and it has a 'Close' column for closing prices

# Step 1: Add month and date columns
start_date = "2022-06-30"  # YYYY-MM-DD format
end_date = "2022-12-12"    # YYYY-MM-DD format

# Generate a date range between start_date and end_date
date_range = pd.date_range(start=start_date, end=end_date)

# Ensure the DataFrame has the same length as the date range
#if len(df) != len(date_range):
    #raise ValueError("The length of the DataFrame does not match the length of the date range.")

# Add the date range as a new column
date_range = pd.date_range(start=start_date, periods=len(df)) # Changed to periods=len(df)
df['date'] = date_range

# Extract month and day from the date column
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

print("\nDataFrame after adding date, month, and day columns:")
print(df.head())

# Step 2: Calculate MACD (Moving Average Convergence Divergence)
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    # Calculate short-term and long-term Exponential Moving Averages (EMA)
    df['EMA_Short'] = df['Close'].ewm(span=short_window, adjust=False).mean()
    df['EMA_Long'] = df['Close'].ewm(span=long_window, adjust=False).mean()

    # Calculate MACD line
    df['MACD'] = df['EMA_Short'] - df['EMA_Long']

    # Calculate Signal line
    df['Signal_Line'] = df['MACD'].ewm(span=signal_window, adjust=False).mean()

    # Calculate MACD Histogram
    df['MACD_Histogram'] = df['MACD'] - df['Signal_Line']

    return df

# Step 3: Calculate RSI (Relative Strength Index)
def calculate_rsi(df, window=14):
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()

    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))

    return df

# Step 4: Calculate Bollinger Bands
def calculate_bollinger_bands(df, window=20, num_std=2):
    df['Rolling_Mean'] = df['Close'].rolling(window=window).mean()
    df['Rolling_Std'] = df['Close'].rolling(window=window).std()

    df['Bollinger_Upper'] = df['Rolling_Mean'] + (df['Rolling_Std'] * num_std)
    df['Bollinger_Lower'] = df['Rolling_Mean'] - (df['Rolling_Std'] * num_std)

    return df

# Apply all preprocessing steps
df = calculate_macd(df)
df = calculate_rsi(df)
df = calculate_bollinger_bands(df)

# Drop NaN values created by rolling calculations
df = df.dropna()

# Display the preprocessed DataFrame
print("\nPreprocessed DataFrame:")
print(df.head())
```

```python
import pandas as pd
import numpy as np

# Assuming `df` is your DataFrame and it has a 'Close' column for closing prices

# Step 1: Add month and date columns
start_date = "2022-06-30"  # YYYY-MM-DD format
end_date = "2022-12-12"    # YYYY-MM-DD format

# Generate a date range between start_date and end_date
date_range = pd.date_range(start=start_date, end=end_date)

# Add the date range as a new column
date_range = pd.date_range(start=start_date, periods=len(df)) # Changed to periods=len(df)
df['date'] = date_range

# Extract month and day from the date column
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

print("\nDataFrame after adding date, month, and day columns:")
print(df.head())

# Step 2: Calculate Moving Averages
def calculate_moving_averages(df, short_window=10, long_window=50):
    # Calculate short-term and long-term Simple Moving Averages (SMA)
    df['SMA_Short'] = df['Close'].rolling(window=short_window).mean()
    df['SMA_Long'] = df['Close'].rolling(window=long_window).mean()
    return df

# Step 3: Calculate Returns
def calculate_returns(df):
    # Calculate daily returns
    df['Daily_Return'] = df['Close'].pct_change()
    return df

# Step 4: Define Criteria for Above-Market and Under-Market Profits/Returns
def define_market_criteria(df):
    # Calculate average daily return as a threshold
    avg_return = df['Daily_Return'].mean()

    # Define above-market and under-market criteria
    df['Market_Condition'] = np.where(
        df['Daily_Return'] > avg_return, 'Above-Market',
        np.where(df['Daily_Return'] < avg_return, 'Under-Market', 'Neutral')
    )
    return df

# Step 5: Calculate MACD (Moving Average Convergence Divergence)
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    # Calculate short-term and long-term Exponential Moving Averages (EMA)
    df['EMA_Short'] = df['Close'].ewm(span=short_window, adjust=False).mean()
    df['EMA_Long'] = df['Close'].ewm(span=long_window, adjust=False).mean()

    # Calculate MACD line
    df['MACD'] = df['EMA_Short'] - df['EMA_Long']

    # Calculate Signal line
    df['Signal_Line'] = df['MACD'].ewm(span=signal_window, adjust=False).mean()

    # Calculate MACD Histogram
    df['MACD_Histogram'] = df['MACD'] - df['Signal_Line']

    return df


# Apply all preprocessing steps
df = calculate_moving_averages(df)
df = calculate_returns(df)
df = define_market_criteria(df)
df = calculate_macd(df)


# Drop NaN values created by rolling calculations
df = df.dropna()

# Display the preprocessed DataFrame
print("\nPreprocessed DataFrame:")
print(df.head())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Define the date range
start_date = "2022-06-30"
end_date = "2022-12-12"

# Load your S&P 500 dataset
# Replace 'path_to_your_file.csv' with the actual path to your dataset
file_path = '/content/AAL.csv'  # Update this path
data = pd.read_csv(file_path)

# Ensure the date column is in datetime format and set it as the index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Filter data for the specified date range
data = data.loc[start_date:end_date]

# Check for different column names for adjusted closing price
adj_close_col = None
for col in ['Adj Close', 'Adjusted Close', 'Adj_Close', 'adjclose']:  # Add other potential names
    if col in data.columns:
        adj_close_col = col
        break

# If a suitable column is found, calculate returns
if adj_close_col:
    # Calculate daily returns using the identified adjusted close column
    data['Daily Return'] = data[adj_close_col].pct_change().dropna()

    # Define the confidence level (e.g., 95%)
    confidence_level = 0.95

    # Calculate the mean and standard deviation of daily returns
    mean_return = data['Daily Return'].mean()
    std_return = data['Daily Return'].std()

    # Calculate VaR using the parametric method (assuming normal distribution)
    VaR = norm.ppf(1 - confidence_level, mean_return, std_return)

    print(f"Value at Risk (VaR) at {confidence_level*100}% confidence level: {VaR:.4f}")

    # Calculate the Expected Shortfall (ES)
    # ES is the average of returns that are worse than the VaR
    returns_below_VaR = data['Daily Return'][data['Daily Return'] <= VaR]
    ES = returns_below_VaR.mean()

    print(f"Expected Shortfall (ES) at {confidence_level*100}% confidence level: {ES:.4f}")
else:
    print("WARNING: File does not have a recognized Adjusted Close column. Skipping...")
```

Portfolio Optimization Models

```python
from scipy.optimize import minimize

# Calculate mean returns and covariance matrix
mean_returns = combined_returns.mean()
cov_matrix = combined_returns.cov()

# Define the objective function (portfolio risk)
def portfolio_risk(weights, cov_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Define constraints (sum of weights = 1)
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

# Define bounds for weights (0 <= weight <= 1)
bounds = tuple((0, 1) for _ in range(len(mean_returns)))

# Initial guess (equal weights)
initial_guess = np.array([1 / len(mean_returns)] * len(mean_returns))

# Optimize the portfolio
result = minimize(portfolio_risk, initial_guess, args=(cov_matrix,), method='SLSQP', bounds=bounds, constraints=constraints)

# Extract optimal weights
optimal_weights = result.x

print("Optimal Portfolio Weights:")
for asset, weight in zip(combined_returns.columns, optimal_weights):
    print(f"{asset}: {weight:.4f}")
```

DDPG Model Building

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
import random
from collections import deque

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Hyperparameters
BUFFER_SIZE = 100000  # Size of the experience replay buffer
BATCH_SIZE = 64       # Mini-batch size for training
GAMMA = 0.99          # Discount factor for future rewards
TAU = 0.005           # Soft update parameter for target networks
ACTOR_LR = 0.001      # Learning rate for the actor network
CRITIC_LR = 0.002     # Learning rate for the critic network
RISK_THRESHOLD = 0.05 # Risk threshold for Value-at-Risk (VaR)

# Environment parameters
STATE_DIM = 10  # Number of features in the state (e.g., Close, MACD, RSI, etc.)
ACTION_DIM = 1  # Number of actions (e.g., position size)
ACTION_HIGH = 1.0  # Maximum value for actions
```

```python
class TradingEnv:
    def __init__(self, data):
        self.data = data  # Financial data (e.g., DataFrame with features)
        self.current_step = 0
        self.max_steps = len(data) - 1

    def reset(self):
        """Reset the environment to the initial state."""
        self.current_step = 0
        return self.data.iloc[self.current_step].values

    def step(self, action):
        """Take a step in the environment."""
        self.current_step += 1
        next_state = self.data.iloc[self.current_step].values
        reward = self._calculate_reward(action)
        done = self.current_step >= self.max_steps
        return next_state, reward, done

    def _calculate_reward(self, action):
        """Calculate reward based on profit/loss and risk."""
        current_price = self.data.iloc[self.current_step]['Close']
        next_price = self.data.iloc[self.current_step + 1]['Close']
        profit = (next_price - current_price) * action
        risk = self._calculate_risk(action)
        reward = profit - risk  # Penalize risky actions
        return reward

    def _calculate_risk(self, action):
        """Calculate risk using Value-at-Risk (VaR)."""
        returns = self.data['Close'].pct_change().dropna()
        var = np.percentile(returns, RISK_THRESHOLD * 100)
        return abs(action) * var  # Risk is proportional to position size
```

```python
def get_actor():
    """Build the actor network."""
    inputs = layers.Input(shape=(STATE_DIM,))
    x = layers.Dense(256, activation="relu")(inputs)
    x = layers.Dense(256, activation="relu")(x)
    outputs = layers.Dense(ACTION_DIM, activation="tanh")(x)  # Outputs between -1 and 1
    outputs = layers.Lambda(lambda x: x * ACTION_HIGH)(outputs)  # Scale to action range
    return tf.keras.Model(inputs, outputs)

def get_critic():
    """Build the critic network."""
    state_input = layers.Input(shape=(STATE_DIM,))
    action_input = layers.Input(shape=(ACTION_DIM,))
    x = layers.Concatenate()([state_input, action_input])
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dense(256, activation="relu")(x)
    outputs = layers.Dense(1)(x)  # Q-value
    return tf.keras.Model([state_input, action_input], outputs)
```

```python
class DDPGAgent:
    def __init__(self):
        """Initialize the DDPG agent."""
        self.actor = get_actor()
        self.critic = get_critic()
        self.target_actor = get_actor()
        self.target_critic = get_critic()

        self.actor_optimizer = tf.keras.optimizers.Adam(ACTOR_LR)
        self.critic_optimizer = tf.keras.optimizers.Adam(CRITIC_LR)

        self.buffer = deque(maxlen=BUFFER_SIZE)  # Experience replay buffer

        # Initialize target networks
        self.target_actor.set_weights(self.actor.get_weights())
        self.target_critic.set_weights(self.critic.get_weights())

    def act(self, state):
        """Select an action using the actor network."""
        state = tf.expand_dims(tf.convert_to_tensor(state), 0)
        action = self.actor(state)[0]
        return action.numpy()

    def remember(self, state, action, reward, next_state, done):
        """Store experience in the replay buffer."""
        self.buffer.append((state, action, reward, next_state, done))

    def update(self):
        """Update the actor and critic networks."""
        if len(self.buffer) < BATCH_SIZE:
            return

        # Sample a mini-batch from the replay buffer
        batch = random.sample(self.buffer, BATCH_SIZE)
        states, actions, rewards, next_states, dones = map(np.array, zip(*batch))

        # Update Critic
        with tf.GradientTape() as tape:
            target_actions = self.target_actor(next_states)
            target_q_values = self.target_critic([next_states, target_actions])
            target_q_values = rewards + GAMMA * (1 - dones) * target_q_values

            q_values = self.critic([states, actions])
            critic_loss = tf.reduce_mean(tf.square(q_values - target_q_values))

        critic_grads = tape.gradient(critic_loss, self.critic.trainable_variables)
        self.critic_optimizer.apply_gradients(zip(critic_grads, self.critic.trainable_variables))

        # Update Actor
        with tf.GradientTape() as tape:
            actions = self.actor(states)
            actor_loss = -tf.reduce_mean(self.critic([states, actions]))

        actor_grads = tape.gradient(actor_loss, self.actor.trainable_variables)
        self.actor_optimizer.apply_gradients(zip(actor_grads, self.actor.trainable_variables))

        # Update target networks
        self._update_target(self.target_actor.variables, self.actor.variables, TAU)
        self._update_target(self.target_critic.variables, self.critic.variables, TAU)

    def _update_target(self, target_weights, weights, tau):
        """Soft update target networks."""
        for (a, b) in zip(target_weights, weights):
            a.assign(tau * b + (1 - tau) * a)
```

```python
def train_ddpg(env, agent, episodes=1000):
    """Train the DDPG agent."""
    for episode in range(episodes):
        state = env.reset()
        total_reward = 0

        while True:
            action = agent.act(state)
            next_state, reward, done = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            agent.update()

            state = next_state
            total_reward += reward

            if done:
                break

        print(f"Episode: {episode + 1}, Total Reward: {total_reward}")
```
